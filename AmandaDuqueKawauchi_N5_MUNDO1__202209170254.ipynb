```python
# Importando bibliotecas 

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
from num2words import num2words
from wordcloud import WordCloud
from faker import Faker
```

```python
# Gerando uma lista de nomes e notas

fake = Faker('pt_br')
n = 10 # Quantidade de nomes e notas geradas, podendo ser alterada conforme a necessidade

# Criação de listas
lista_de_nomes = []
lista_de_notas = []
```

```python
# Criando arquivo de texto e adicionando nomes e notas
""" Manipular arquivos de texto é importante para realizar análises das palavras utilizadas nos textos, 
o que pode formar bancos de dados para visualização de vários tipos de instrumentos de análise, 
como a nuvem de palavra """

with open("alunos.txt", "w+") as alunos:
  for i in range(n):
    nome = fake.name()
    ponto = str(random.randint(0,10))
    resultado = nome + " " + ponto + "\n"
    alunos.write(resultado)
```

```python
# Manipulando dados do arquivo de texto criado

with open("alunos.txt", "r") as dados:
  for i in range(n):
    leitura_de_dados = dados.readline()
    espaços = leitura_de_dados.split(' ')
    leitura_de_notas = int(espaços[-1])
    lista_de_notas.append(leitura_de_notas)
    lista_de_nomes_ext = num2words(leitura_de_notas, lang = 'pt_BR')
    lista_de_nomes.append(lista_de_nomes_ext)
    lista_de_nomes_str = (" ").join(lista_de_nomes)
```

```python
# Criando o Histograma
""" A criação de um histograma é importante para ver a 
distribuição da frequência dos dados"""

plt.title('Histograma de Pontuações')
plt.xlabel('Pontuações')
plt.ylabel('Probabilidade')
plt.hist(lista_de_notas, facecolor='blue', density=True, alpha=0.75)
plt.grid(True)
plt.show()
```

```python
# Criando uma nuvem de palavras
""" A visualização da nuvem de palavras é importante para transpor para uma
analise grafica a frequência das palavras mais usadas """

import wordcloud
wordcloud = WordCloud().generate(lista_de_nomes_str)
wordcloud = WordCloud(background_color='#1c1c1c', width=800,height=400).generate(lista_de_nomes_str)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
wordcloud.to_file("Nuvem_de_palavras.png")
plt.close()
```
